## Data Splitting for Training and Testing

```python
import numpy as np
```

```python
def split_data(data, ratio):
    np.random.seed(42)
    splitted = np.random.permutation(len(data))
    print(splitted)
    testing_size = int(len(data) * (ratio))
    testing_index = splitted[:testing_size]
    training_index = splitted[testing_size:]
    return data.iloc[training_index], data.iloc[testing_index]
```

```python
training_set , testing_set = split_data(housing, 0.2)
```

```python
print(len(training_set))
```

```python
print(len(testing_set))
```

```python
from sklearn.model_selection import train_test_split
training_set, testing_set = train_test_split(housing, test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing['CHAS']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
housing = strat_train_set.copy()
```

## Missing Attributes

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer( strategy='median')
imputer.fit(housing)
```

```python
imputer.statistics_
```

```python
x = imputer.transform(housing)
```

```python
housing_tr = pd.DataFrame(x,columns=housing.columns)
```

```python
housing_tr.describe()
```

## Scikit Learn Design

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipes = Pipeline([('imputer', SimpleImputer(strategy='median')),('std_scaler',StandardScaler())])
```

```python
housing_num_tr = pipes.fit_transform(housing_tr)
```

```python
housing_num_tr
```

## Selecting model for Real Estate

```python
# from sklearn.linear_model import LinearRegression
# model = LinearRegression()
# from sklearn.tree import DecisionTreeRegressor
# model = DecisionTreeRegressor()
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(housing_num_tr, housing_labels)
```

```python
data = housing.iloc[:5]
```

```python
labels = housing_labels.iloc[:5]
```

```python
prepared_data = pipes.transform(data)
```

```python
model.predict(prepared_data)
```

```python
labels
```

## Calculation 

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels,housing_predictions)
rmse = np.sqrt(mse)
```

```python
mse
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,housing_num_tr,housing_labels, scoring="neg_mean_squared_error", cv = 10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("Std Deviation: ", scores.std())
```

```python
print_scores(rmse_scores)
```

## Saving the Model

```python
from joblib import dump, load
dump(model, 'data.joblib')
```

## Testing Model

```python
x_test = strat_test_set.drop("MEDV",axis=1)
y_test = strat_test_set["MEDV"].copy()
x_test_prepared = pipes.transform(x_test)
final_predictions = model.predict(x_test_prepared)
final_mse = mean_squared_error(y_test,final_predictions)
final_rmse = np.sqrt(final_mse)
```
